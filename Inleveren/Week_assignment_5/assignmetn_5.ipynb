```python
#import important libraries
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
#do a html request from the feyenoordwebsite using the beautifulsoup 
req = requests.get("https://www.feyenoord.nl/teams/feyenoord-1/selectie")
html = req.text
page = BeautifulSoup(html, "lxml")
```

```python
#Get the page title first to see if correct page
page.select("title")[0].get_text().strip()
```

```python
#create a function named get_player the input should be an html link and the return should be a list of players.
def get_player(html): 
    #beautifulsoup the inputted html (mooie zin)
    soup = BeautifulSoup(html, "lxml")
    #create a empty players list, which should be returned at the end of the function
    players = []
    #Select each double class named .person.person__link
    players_list = page.select(".person.person__link")
    #print the amount of players found, to check if it went well 
    print(f"found {len(players_list)} players")
    
    #loop through the players list
    for player in players_list:
        #select the person__number class, choose the first element. Get the text(delete the html stuff) and strip the answer and repeat for
        #name and function
        number = player.select(".person__number")[0].get_text().strip()
        name = player.select(".person__name")[0].get_text().strip()
        function = player.select(".person__function")[0].get_text().strip()
        #For the player variable (which is the selected .person.person__link) get the attr href (link)
        url = player.attrs["href"].strip()
        
        #Add the variables above to the item variable
        item = {
            "name" : name,
            "function": function,
            "number" : number,
            "link" : url,
           
        }#Add this variable to the list of players 
        players.append(item)
     

    #return the players list
    return players
```

```python
#since the players list is just a local variable in the function, we need to put it in a global variable named.... players :) 
players = get_player(html)
```

```python
#Create a new dataframe with the players variable
df = pd.DataFrame(players)
```

```python
#create a csv and write the dataframe to it.
df.to_csv("Selectie_feyenoord.csv","w")
```

```python
#New empty list with players_details
players_details = [ ]
```

```python
#new function with input html
def get_details(html):
    #Same steps as above
    req = requests.get(f"{html}")            
    html = req.text
    page = BeautifulSoup(html, "lxml")
    #now create the players_details_list (long name iknow) with the element with a content class 
    players_details_list = page.select(".content")
    #Print the length of this list
    print(f"the list contains {len(players_details)} players")


    #Loop through this list 
    for detail in players_details_list:
        #select the element with the player_value class and select the first, second or third element and add it to the correct variable
        birth_date = detail.select("dd.player__value")[0].get_text().strip()
        nation = detail.select(".player__value")[1].get_text().strip()
        contract_till = detail.select(".player__value")[2].get_text().strip()
        #IF the player has made his debut in the first team of feyenoord, add this date to a variable. if not, add NAN
        if "Debuut Feyenoord 1" in html:
            debut_feyenoord = detail.select(".playerDebut__date")[0].get_text().strip()
        else:
            debut_feyenoord = "NAN"
        #IF the player has made his debut in the national team, add this date to a variable. if not, add NAN
        if "Internationaal debuut" in html:
            International_debut = detail.select(".playerDebut__date")[1].get_text().strip()
        else:
            International_debut = "NAN"
        #IF the player has played at another club, add this date to a variable. if not, add NAN
        if "Voorgaande clubs" in html:
            previous_club = detail.select(".playerPreviousClubs__team")[-1].get_text().strip()
        else: 
            previous_club = "NAN"

         
        #The lovely items variable
        items = {
        "birth_date" : birth_date,
        "nation": nation,
        "contract_till" : contract_till,
        "debut_feyenoord" : debut_feyenoord,
        "previous_club": previous_club,
        "International_debut" : International_debut

        }
        #Adding this to the list...
        players_details.append(items)
    #Returning the list of player detailsss
    return players_details

```
